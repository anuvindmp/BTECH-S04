```python
def addEdge(node1, node2):
    if node1 not in graph:
        graph[node1] = []
    if node2 not in graph:
        graph[node2] = []
    if node2 not in graph[node1]:
        graph[node1].append(node2)
```

```python
#graph creation
addEdge('A', 'B')
addEdge('A', 'C')
addEdge('B', 'D')
addEdge('B', 'E')
addEdge('C', 'D')
addEdge('D', 'E')
addEdge('D', 'A')

for node in graph:
  graph[node] = list(set(graph[node]))

print("Graph : \n", graph)
print("\nVertices : ", list(graph.keys()))
print("\nEdges of the graph are : ")
for node, neighbors in graph.items():
    for neighbor in neighbors:
      print(f"({node}, {neighbor})")
```

```python
graph = {}

def addEdge(node1, node2, weight):
    if node1 not in graph:
        graph[node1] = {}
    if node2 not in graph:
        graph[node2] = {}
    if node2 not in graph[node1]:
        graph[node1][node2] = weight
```

```python
#graph creation
addEdge(0,1,3)
addEdge(0,3,8)
addEdge(0,4,7)
addEdge(1,2,1)
addEdge(1,3,4)
addEdge(2,3,2)
addEdge(3,4,3)

print("Graph :", graph)
print("\nVertices : ", list(graph.keys()))
print("\nEdges of the graph are : ")
for node, neighbors in graph.items():
    for neighbor in neighbors:
      print(f"({node}, {neighbor})")
```
