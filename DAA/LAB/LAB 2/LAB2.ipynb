```python
def removeDuplicates(nums):
        a = []
        for i in range(len(nums)):
            if nums[i] != nums[i-1]:
                a.append(nums[i])
        nums = a
        return nums
    
A = [1,1,2]
print(removeDuplicates(A))
```

```python
def splitArray(nums):
    Low, High, chk= [], [], nums[0]
    for i in nums:
        if i < chk:
            Low.append(i)
        else:
            High.append(i)
    return Low,High

nums = [5,7,2,3,5,1,2,9,3]
a,b = splitArray(nums)
print(f"array : {nums} \nlow : {a} \nhigh : {b}")
```

```python
def merge(A, B):
    res = []
    i,j = 0, 0
    n = len(A) + len(B)
    while i < len(A) and j < len(B):
        if A[i] < B[j]:
            res.append(A[i])
            i+=1
        else:
            res.append(B[j])
            j+=1
    if i<len(A):
        res+=A[i:]
    if j<len(B):
        res+=B[j:]

    return res

A = [1,3,6,7]
B = [2,4,5,8]
print(merge(A,B)) 
```

```python
def sortTheStudents(score, k):
        output = []
        x = [i[k] for i in score]

        graph = {}
        j = 0
        for i in x:
            graph[i] = j
            j+=1
        
        x.sort(reverse = True)
        
        for i in x:
            key = graph[i]
            output.append(score[key])

        return output

score = [[10,6,9,1],[7,5,11,2],[4,8,3,15]]
print("Before sorting : ",score)
print("\nAfter sorting : ",sortTheStudents(score,2))
```
