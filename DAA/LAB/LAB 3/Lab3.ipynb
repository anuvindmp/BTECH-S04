```python
def sum_(n):
    if n == 1:
        return 1
    else:
        return n + sum_(n-1)

sum_(5)
```

```python
def prdct(n):
    if n == 1:
        return 1
    else :
        return n*prdct(n-1)

prdct(5)
```

```python
def fib(n):
    if n == 1:
        return 0
    elif n == 2:
        return 1
    else:
        return fib(n-1) + fib(n-2)
    
fib(5)
```

```python
def power(x,y):
    if y == 0:
        return 1
    elif y == 1:
        return x
    else:
        return x*pow(x,y-1)
    
power(2,10)
```

```python
def natural(n):
    if n == 0:
        return False
    else:
        natural(n-1)
        print(n, end = ", ")
        

natural(4)
```

```python
def naturalrev(n):
    if n == 0:
        return False
    else:
        print(n, end = ",")
        naturalrev(n-1)

naturalrev(4) 
```

```python
def GCD(a,b):
    if b == 0:
        return a
    else:
        return GCD(b, a%b)
    
GCD(10,20)
```

```python
def printArray(arr):
    if arr == []:
        return
    else:
        print(arr[0], end=' ')
        printArray(arr[1:])

arr = [1,2,3]
printArray(arr)
```

```python
def revPprintArray(arr):
    if arr == []:
        return
    else:
        revPprintArray(arr[1:])
        print(arr[0], end = " ")

arr = [1,2,3]
revPprintArray(arr)
```

```python
def reverse(n, rev=0):
    if n == 0:
        return rev
    x = n % 10
    rev = rev * 10 + x
    return reverse(n // 10, rev)
reverse(1234)
```

```python
def check(arr):
    if len(arr) < 2:
        return True
    elif arr[0]>arr[1]:
        return False
    else:
        return check(arr[1:])
        
    
arr = [1,2,4,3]
arr2 = [1,2,3]
print(check(arr))
print(check(arr2))
```
